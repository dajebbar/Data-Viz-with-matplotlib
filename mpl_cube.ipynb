# Plotting a Simple Line Graph
_Plot a simple line graph using Matplotlib, 
and then customize it to create a more informative data visualization._

```python
import matplotlib.pyplot as plt

cubes = [1, 8, 27, 64, 125]

fig, ax = plt.subplots()
ax.plot(cubes)

plt.show()
```

# Changing the Label Type and Line Thickness
_Adjust features of a visualization._

```python
cubes = [1, 8, 27, 64, 125]
fig, ax = plt.subplots()
ax.plot(cubes, linewidth=3)

# Set chart title and label axes.
ax.set_title("Cube Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Cube of Value", fontsize=14)

# Set size of tick labels.
ax.tick_params(axis='both', labelsize=14)

plt.show()
```

# Correcting the Plot
_Plotting data correctly._

```python
values = [1, 2, 3, 4, 5]
cubes = [1, 8, 27, 64, 125]
fig, ax = plt.subplots()
ax.plot(values, cubes, linewidth=3)

# Set chart title and label axes.
ax.set_title("Cube Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Cube of Value", fontsize=14)

# Set size of tick labels.
ax.tick_params(axis='both', labelsize=14)

plt.show()
```

# Using Built-in Styles
_Make visualizations appealing._

```python
values = [1, 2, 3, 4, 5]
cubes = [1, 8, 27, 64, 125]

plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()

ax.plot(values, cubes, linewidth=3)

# Set chart title and label axes.
ax.set_title("Cube Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Cube of Value", fontsize=14)

# Set size of tick labels.
ax.tick_params(axis='both', labelsize=14)

plt.show()
```

# Plotting a Series of Points with scatter()
_Plot and style points._

```python
values = [1, 2, 3, 4, 5]
cubes = [1, 8, 27, 64, 125]

plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()

ax.scatter(values, cubes, s=100)

# Set chart title and label axes.
ax.set_title("Cube Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Cube of Value", fontsize=14)

# Set size of tick labels.
ax.tick_params(axis='both', labelsize=14)

plt.show()
```
